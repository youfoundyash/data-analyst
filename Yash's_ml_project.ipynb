## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Answer Here.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

* Null Hypothesis (H0): Stores with and without promotions do not significantly differ in sales.
* Alternative Hypothesis (H1): Stores with and without promotions have significantly different sales.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
import scipy.stats as stats

# creating two groups based upon weather the store has promotion or not
pg = new_data[new_data['PromoInterval'] == 0]['Sales']    # pg =  promotion group
npg = new_data[new_data['PromoInterval'] != 0]['Sales']   # npg = non promotion group

# performing T-test to compare the means of two groups

t_statistic , p_value = stats.ttest_ind(pg,npg)

# defining the significance level

alpha = 0.05

# paking decision based on the p-level

if p_value < alpha :
  print("Reject the null hypothesis: There is a significant difference in sales between stores with and without promotions.")
else :
  print("Fail to reject the null hypothesis: There is no significant difference in sales between stores with and without promotions.")
````

##### Why did you choose the specific statistical test?

Since the t-test compares the means of two independent groups and is a parametric test, that's why I went with it. Given that the two groups are independent and the data is normally distributed, the t-test is suitable in this scenario.

* Alternative Hypothesis (H1): Sales in stores with various assortment types differ significantly from one another.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value

# creating three groups acoording to the Assortment type

a_a = new_data[new_data['Assortment'] == 'a']['Sales']
a_b = new_data[new_data['Assortment'] == 'b']['Sales']
a_c = new_data[new_data['Assortment'] == 'c']['Sales']

# performing ANOVA test to test the mean of all the three group
f_statistic , p_value = stats.f_oneway(a_a , a_b , a_c )

# defining the significance level

alpha = 0.05

# mapking decision based on p-value
if p_value < alpha :
    print("Reject the null hypothesis: There is a significant difference in sales between stores of different assortment types.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in sales between stores of different assortment types.")
````

##### Why did you choose the specific statistical test?

Since the ANOVA test compares the means of three or more independent groups, it is a parametric test that I selected. Given that the three groups are independent and the data is normally distributed, the ANOVA test is suitable in this scenario.
